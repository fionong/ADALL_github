```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Replace this with the raw URL of your GitHub file
github_raw_url = 'https://raw.githubusercontent.com/fionong/ADALL_github/refs/heads/main/laptop_prices_2024_sgd_TL.csv'
try:
    df = pd.read_csv(github_raw_url)
    print("Successfully loaded data from GitHub!")
    display(df.head())
except Exception as e:
    print(f"Error loading data: {e}")
    print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")
```

```python
#plot each df graph
df.plot(kind='bar')
plt.show()

#plot each df col graph
df.plot(kind='bar', subplots=True)
plt.show()
```

```python
df.info()
df.describe()
```

```python
import numpy as np
from io import StringIO

#................
#Generate a full dataset profile
#----------------

buffer = StringIO()

#dtypes
buffer.write(df.dtypes.to_string())
buffer.write('\n\n')

#missing values
buffer.write(df.isnull().sum().to_string())
buffer.write('\n\n')

#correlation
buffer.write(df.select_dtypes(include=np.number).corr().to_string())
buffer.write('\n\n')

#numeric describe
buffer.write(df.describe().to_string())
buffer.write('\n\n')

#categorical describe
buffer.write(df.describe(include=object).to_string())
buffer.write('\n\n')

buffer.getvalue()
```

```python
try:
  df_numerical = df.select_dtypes(exclude='bject')
  df_numerical.describe(include='object')
except:
  print(f"No catergorical columns found")
```

```python
from google.colab import userdata
from openai import OpenAI

# Load key from Google Colab Secrets
api_key = userdata.get('OPENAI_API_KEY')


client = OpenAI(
    # This is the default and can be omitted
    api_key=api_key,
)

# Convert the first few rows to a string to send to OpenAI
data_preview = df.head(10).to_csv(index=False)
print(data_preview)

response = client.responses.create(
    model="gpt-4o-mini",
    instructions="You are an expert data scientist with extensive knowledge of predictive analysis and linear regression.",
    input=f"Dataset: Laptop specifications and prices (1000 samples, 13 features, 1 classes)\nHere are the first 10 rows of the dataset:\n{data_preview}]",
)
print(response.output_text)
```
